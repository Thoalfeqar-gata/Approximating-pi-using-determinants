# Approximating pi using determinants
The goal of this notebook is to approximate the value of pi by calculating the area of a unit circle using the following determinant formula:



First, let us make the circle:

```python
import numpy as np
from matplotlib import pyplot as plt

n_edges = 1000
X, Y = np.zeros(n_edges), np.zeros(n_edges)
for i in range(n_edges):
    angle_in_radians = 360 * i/n_edges * np.pi/180
    x = np.cos(angle_in_radians)
    y = np.sin(angle_in_radians)
    
    X[i] = x
    Y[i] = y

plt.fill(X, Y)
plt.title(f'Approximate unit circle ({n_edges}) edges')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

    
```

Now, let us calculate its area using the determinant formula:

```python
area = 0
for i in range(n_edges - 1):
    matrix = np.array([
        [X[i], Y[i]],
        [X[i+1], Y[i+1]]
    ])
    area += 1/2 * np.linalg.det(matrix)

area, np.allclose(area, np.pi)
```

From the above result, if we set the n_edges parameter to a high enough value, the area will become a better approximation of the value of pi!
